# 3. Orbitas planetarias:  ́ En el espacio, la orbita de un cuerpo alrededor de otro (como un planeta alrededor del Sol), 
# no necesariamente es circular. En general, toma la forma de una elipse, con el cuerpo a veces mas cerca y otras más lejos. 
# Si tenemos la distancia l1 de maxima aproximación de un planeta al Sol (su perihelio), y su velocidad lineal v1 en
# el perihelio, entonces cualquier otra propiedad de la orbita se puede calcular a partir de estas dos cantidades de la siguiente manera:

#### b) Dados los valores de 
$v_1$ , $l_1$, y $l_2$; 
#### otros parametros de la  ́órbita se obtienen mediante formulas simples que pueden derivarse de las leyes de Kepler y del hecho de que la orbita es una elipse:  ́

Semieje mayor: $ a = \frac{1}{2}(l_1+l_2) $

Semieje menor: $ b = \sqrt{l_1l_2}$

Perıodo orbital: $ T =\frac{2\pi ab}{l_1v_1} $ 

Excentricidad orbital:$ e =\frac{l_2-l_1}{l_2+l_1} $
#### Escribe un programa que le pida al usuario que ingrese la distancia al Sol y la
#### velocidad en el perihelio; para que calcule e imprima las cantidades 
$l_2$, $v_2$, $T$ y $e$.

```python
M = 1.9891e30 #kg es la masa del Sol y 
G = 6.6738e-11 #m3 kg−1-s−2 es la constante gravitatoria de Newton.

def datos_orbita(ell1, v1):
     #Aquí se asigna variables a ciertos datos para que al resolver la cuadrática esta sea más fácil de escribrir.
    a = 1
    b = -(2 * G * M) / (ell1 * v1)
    c = -(v1**2 - (2 * G * M) / ell1)
    v2_1 = (-b + sqrt(b**2 - 4 * a * c)) / (2 * a)
    v2_2 = (-b - sqrt(b**2 - 4 * a * c)) / (2 * a)

    # Selecciona la raíz más pequeña para v2
    v2 = min(v2_1, v2_2)

    #Aquí se calcula la distancia en el afelio l2
    ell2 = (ell1 * v1) / v2

    #Aquí se calcula el semieje mayor a, semieje menor b, periodo T y excentricidad e
    a = 0.5 * (ell1 + ell2)
    b = sqrt(ell1 * ell2)
    T = (2 *pi * a * b) / (ell1 * v1)
    e = (ell2 - ell1) / (ell2 + ell1)
    return ell2, v2, T, e

#c) A continuación se hará la prueba con los datos de la tierra:
# l1 = 1.4710e11 m
# v1 = 3.0287e4 m/s

ell_1 = float(input("Ingrese la distancia al Sol en el perihelio (en metros): "))
v1 = float(input("Ingrese la velocidad en el perihelio (en metros por segundo): "))

ell2, v2, T, e = datos_orbita(ell_1, v1)
print(f"Distancia en el afelio (l2): {ell2:.2e} metros")
print(f"Velocidad en el afelio (v2): {v2:.2f} metros/segundo")
print(f"Periodo orbital (T): {T:.2f} segundos")
print(f"Excentricidad orbital (e): {e:.4f}")


#Al eejcutarse las pruebas, se encontró que T=31543060.21 segundos que equivale a un año
  
```
